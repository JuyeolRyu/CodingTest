```python
def solution(p):
    answer = ''
    cnt = 0
    index = 0

    #1.입력이 빈 문자열인 경우
    if(p == ''):
        return p
    #2.문자열 분리
    index = split_str(p)
    u = p[0 :index+1]
    v = p[index+1 : len(p)]
    #3.올바른 문자열인지 판단
    if(is_correct(u)):#u가 올바른 문자열인 경우
        answer = u + solution(v)
    else:#u가 올바른 문자열이 아닌 경우
        answer = correction(u,v)

    return answer

def is_correct(u):
    cnt = 0
    for i in range(len(u)):#')'가 먼저 나오는 경우는 틀린 겨우 false 반환
        if(u[i]=='('):
            cnt += 1
        elif(u[i]==')'):
            cnt -= 1

        if(cnt<0):
            return False
    return True

def split_str(p):#문자열 u,v로 분할
    cnt = 0
    index = 0
    for i in range(len(p)):
        if(p[i]=='('):
            cnt += 1
        elif(p[i]==')'):
            cnt -= 1

        if(cnt == 0):
            index = i
            break
    return index
def correction(u,v):#올바른 문자열로 고치기
    result = []
    result.append('(')
    result.append(solution(v))
    result.append(')')
    for i in range(1,len(u)-1):
        if(u[i] == '('):
            result.append(')')
        else:
            result.append('(')

    tmp = ''.join(result)
    return tmp
```
