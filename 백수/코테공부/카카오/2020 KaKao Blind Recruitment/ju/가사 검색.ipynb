```python
class Node(object):
    def __init__(self,key):
        self.key = None
        self.data = []
        self.children = {}
class Trie(object):
    def __init__(self):
        self.head = Node(None)
        
    def insert(self,word):
        current_node = self.head
        
        for char in word:
            if char not in current_node.children:
                current_node.children[char] = Node(char)
            current_node.data.append(word)
            current_node = current_node.children[char]
        
    
    def search(self,query):
        current_node = self.head
        
        for char in query:
            if char == '?':
                cnt = 0
                for i in current_node.data:
                    if len(query) == len(i):
                        cnt += 1
                return cnt
            else:
                if char in current_node.children:
                    current_node = current_node.children[char]
                else:
                    return 0
        
        return 0
                
def solution(words, queries):
    answer = []
    query_dict = {}
    trie = Trie()
    reverse_trie = Trie()
    
    for word in words:
        trie.insert(word)
        reverse_trie.insert(word[::-1])
        
    for query in queries:
        tmp = 0

        if query not in query_dict:
            if query[0] == '?':
                tmp = reverse_trie.search(query[::-1])
                answer.append(reverse_trie.search(query[::-1]))
            else:
                tmp = trie.search(query)
                answer.append(trie.search(query))

            query_dict[query] = tmp
        else:
            answer.append(query_dict[query])

    return answer
```
